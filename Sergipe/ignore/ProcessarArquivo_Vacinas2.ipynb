```python
list_columns = [
    'Municipio',
    'Remessa1_Enviadas',
    'Remessa1_Aplicadas',
    'Remessa1_Cobertura',
    'Remessa2_Enviadas',
    'Remessa2_Aplicadas',
    'Remessa2_Cobertura',
    'Total_Enviadas',
    'Total_Aplicadas'
]
```

```python
final_columns = [
    'Cd_Municipio',
    'Nome_Municipio',
    'Remessa1_Enviadas',
    'Remessa1_Aplicadas',
    'Remessa2_Enviadas',
    'Remessa2_Aplicadas',
    'Total_Aplicadas'
]
```

```python
int_columns = [final_columns[i] for i in [2,3,4,5,6]]
```

```python
data_final = pd.read_excel("Doses-aplicadas-08.02.xlsx",skiprows=1,names=list_columns)
data_final = data_final.head(-1)
data_final[['Cd_Municipio', 'Nome_Municipio']] = data_final['Municipio'].str.split(' - ',expand=True)
data_final = data_final[final_columns]
data_final = data_final.fillna(0)
data_final[int_columns] = data_final[int_columns].astype(int)
data_final
```

```python
data_final.to_csv("relatorio_vacina_sergipe.csv", index=False)
```
