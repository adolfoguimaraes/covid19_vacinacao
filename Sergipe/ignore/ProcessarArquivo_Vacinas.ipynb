```python
import pandas as pd
```

```python
names_columns = [
    'Municipio',
    'TrabalhadorSaude_Total',
    'Grupo1_DosesEnviadas',
    'Grupo1_DosesAplicadas',
    'Grupo1_Cobertura',
    'Grupo2_DosesEnviadas',
    'Grupo2_DosesAplicadas',
    'Grupo2_Cobertura',
    'Grupo3_DosesEnviadas',
    'Grupo3_DosesAplicadas',
    'Grupo3_Cobertura',
    'Remessa1_Enviadas',
    'Remessa1_Aplicadas',
    'Remessa1_Cobertura',
    'Remessa2_Enviadas',
    'Remessa2_Aplicadas',
    'Remessa2_Cobertura',
    'TrabalhadorSaude_Enviadas',
    'TrabalhadorSaude_Aplicadas',
    'TrabalhadorSaude_Cobertura',
    'Total_Aplicadas'
]
```

```python
data_ultimo = pd.read_excel("Doses-aplicadas03fev21.xlsx", skiprows=1, names=names_columns, convert_float=False)
data_ultimo = data_ultimo.head(-1)
data_ultimo[['Cd_Municipio', 'Nome_Municipio']] = data_ultimo['Municipio'].str.split(' - ',expand=True)
data_ultimo = data_ultimo[final_columns]
data_ultimo = data_ultimo.fillna(0)
data_ultimo[final_columns[2:]] = data_ultimo[final_columns[2:]].astype(int)
data_ultimo
```
