```python
import pandas as pd
import glob
from datetime import datetime
```

```python
files = glob.glob("output/historico/*.csv")

data_list = []

for file in files:
    date_ = file.split("/")[-1].split(".")[0]
    date_ = datetime.strptime(date_, '%Y%m%d').date()
    
    data_ = pd.read_csv(file)
    data_ = data_[['enviadas_dose1','aplicadas_dose1','enviadas_dose2','aplicadas_dose2']]
    data_sum = data_[['enviadas_dose1','aplicadas_dose1','enviadas_dose2','aplicadas_dose2']].sum()
    list_ = [date_,data_sum['enviadas_dose1'],data_sum['aplicadas_dose1'],data_sum['enviadas_dose2'],data_sum['aplicadas_dose2']]
    data_list.append(list_)

columns_names = ['data','enviadas_dose1','aplicadas_dose1','enviadas_dose2','aplicadas_dose2']
data_final = pd.DataFrame(data_list,columns=columns_names)
data_final.sort_values(by=['data'],inplace=True)
data_final
```

```python
data_final.to_csv("output/historico_sergipe.csv",index=False)
```
