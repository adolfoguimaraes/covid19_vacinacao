```python
import pandas as pd
import numpy as np
```

```python
caso_full = pd.read_csv('data_general/caso_full.csv')
```

```python
caso_full.columns
```

```python
caso_full_sergipe = caso_full.query('state == "SE" and city.isnull()', engine='python')
caso_full_sergipe
```

```python
caso_full_selected = caso_full_sergipe[['date','last_available_confirmed','last_available_deaths','new_confirmed','new_deaths']]
caso_full_selected
```

```python
caso_full_selected['mean_confirmed'] = caso_full_selected['new_confirmed'].rolling(window=7).mean()
caso_full_selected['mean_deaths'] = caso_full_selected['new_deaths'].rolling(window=7).mean()
```

```python
caso_full_selected = caso_full_selected.fillna(0)
caso_full_selected
```

```python
caso_full_selected['pct_change_confirmed'] = caso_full_selected['mean_confirmed'].pct_change(14)
caso_full_selected['pct_change_deaths'] = caso_full_selected['mean_deaths'].pct_change(14)
caso_full_selected = caso_full_selected.fillna(0)
caso_full_selected
```

```python
caso_full_selected.reset_index(inplace=True,drop=True)
caso_full_selected
```

```python
caso_full_selected.to_csv("output/sergipe_historico.csv", index=False)
```
